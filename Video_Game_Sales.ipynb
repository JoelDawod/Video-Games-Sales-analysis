```python
fdata = data2.copy()
```

```python
fdata.head()
```

###Which titles sold the most worldwide?

```python
fdata.groupby('title')['total_sales'].sum().sort_values(ascending=False)
#the most sold title is "Grand Theft Auto V"
```

###Which year had the highest sales? Has the industry grown over time?

```python
#the year wit hthe highest sales is 2008
#the industry grow over time but the sales started decreasing after 2011 and rapidly after 2014
fdata['year-month'] = fdata['release_date'].dt.to_period('Y').astype(str)
salesByDate = fdata.groupby('year-month')['total_sales'].sum()
salesByDate.sort_index(inplace=True)
plt.figure(figsize=(10,5))
plt.plot(salesByDate.index, salesByDate.values)
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Year')
plt.xticks(rotation=90)
plt.show()

salesByDate.idxmax()
```

###Do any consoles seem to specialize in a particular genre?

```python
#yes consoles seem to specialize in a particular genra
#as an example: PS4 seem to specializa in shooter games
genre_sales_by_console  = fdata.groupby(['console', 'genre'])['total_sales'].sum().unstack(fill_value=0)
genre_distribution = genre_sales_by_console.div(genre_sales_by_console.sum(axis=1), axis=0) * 100
dominant_genre = genre_distribution.idxmax(axis=1)
print(dominant_genre)
dominant_genre['PS4']
```

###What titles are popular in one region but flop in another?

```python
reigonal_sales = fdata.copy()
reigonal_sales = reigonal_sales.groupby('title')[['total_sales','na_sales', 'jp_sales', 'pal_sales', 'other_sales']].sum().sort_values(by='total_sales', ascending=False)
reigonal_sales.drop('total_sales', axis=1, inplace=True)
reigonal_sales['max'] = reigonal_sales[['na_sales', 'jp_sales', 'pal_sales', 'other_sales']].idxmax(axis=1)
reigonal_sales['min'] = reigonal_sales[['na_sales', 'jp_sales', 'pal_sales', 'other_sales']].idxmin(axis=1)

labels = {'na_sales':'NA', 'jp_sales':'JP', 'pal_sales':'PAL', 'other_sales':'Other'}
reigonal_sales.replace({'max':labels}, inplace=True)
reigonal_sales.replace({'min':labels}, inplace=True)

reigonal_sales['maxV'] = reigonal_sales[['na_sales', 'jp_sales', 'pal_sales', 'other_sales']].max(axis=1)
reigonal_sales['minV'] = reigonal_sales[['na_sales', 'jp_sales', 'pal_sales', 'other_sales']].min(axis=1)
reigonal_sales['diff'] = reigonal_sales['maxV'] - reigonal_sales['minV']
reigonal_sales['var'] = reigonal_sales[['na_sales', 'jp_sales', 'pal_sales', 'other_sales']].var(axis=1)
reigonal_sales.sort_values(by='diff', ascending=False, inplace=True)
reigonal_sales.head()
```
